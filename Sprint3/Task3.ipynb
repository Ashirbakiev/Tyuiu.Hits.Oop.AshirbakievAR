<b>Описание задачи:</b>
Создать базовый класс Student в C#, который будет представлять информацию о
студентах. На основе этого класса разработать 2-3 производных класса,
демонстрирующих принципы наследования и полиморфизма. В каждом из классов
должны быть реализованы новые атрибуты и методы, а также переопределены
некоторые методы базового класса для демонстрации полиморфизма.

<b>Требования к базовому классу Student:</b><br>
• Атрибуты: Имя (Name), Возраст (Age), Курс (Course).<br>
• Методы:
o GetInfo(): метод для получения информации о студенте в виде строки.
o Study(): метод для вывода сообщения о том, что студент учится.
o TakeExam(): метод для вывода сообщения о сдаче экзамена.
<b>Требования к производным классам:</b>
1. Студент бакалавриата (BachelorStudent): Должен содержать дополнительные
атрибуты, такие как Специальность (Specialty). Метод Study() должен быть
переопределен для добавления информации о специальности при изучении
предмета.
2. Студент магистратуры (MasterStudent): Должен содержать дополнительные
атрибуты, такие как Научный руководитель (ScientificAdvisor).
Метод TakeExam() должен быть переопределен для добавления
информации о научном руководителе при сдаче экзамена.
3. Аспирант (PhDStudent) (если требуется третий класс): Должен содержать
дополнительные атрибуты, такие как Тема диссертации (DissertationTopic).
Метод Study() должен быть переопределен для добавления информации о
теме диссертации при изучении предмета.<br>

<b>Дополнительное задание</b><br>
Добавьте к сущестующим классам (базовыму и производным 3-4 атрибута и метода) и реализуйте полиморфизм с перекрытием и прегегрузкой методов, а также generic классы

<h2 style="color:DodgerBlue">Реализация:</h2>

```
class Student
{
    public string Name { get; set; }
    public int Age { get; set; }
    public string Course { get; set; }
    public double GPA { get; set; }
    public string University { get; set; }

    // Дополнительные атрибуты
    public string Email { get; set; }
    public string PhoneNumber { get; set; }

    public Student(string name, int age, string course, double gpa, string university, string email, string phoneNumber)
    {
        Name = name;
        Age = age;
        Course = course;
        GPA = gpa;
        University = university;
        Email = email;
        PhoneNumber = phoneNumber;
    }

    // Полиморфизм: перегрузка метода GetInfo
    public virtual string GetInfo()
    {
        return $"Имя: {Name}, Возраст: {Age}, Курс: {Course}, Средний балл: {GPA}, Университет: {University}, Email: {Email}, Телефон: {PhoneNumber}";
    }

    // Полиморфизм: перегрузка метода GetInfo
    public string GetInfo(string format)
    {
        if (format == "short")
        {
            return $"Имя: {Name}, Курс: {Course}";
        }
        return GetInfo();
    }

    // Полиморфизм: перекрытие методов
    public virtual void Study()
    {
        Console.WriteLine($"{Name} учится в университете {University}.");
    }

    public virtual void TakeExam()
    {
        Console.WriteLine($"{Name} сдает экзамен.");
    }

    // Дополнительные методы
    public void AttendLecture()
    {
        Console.WriteLine($"{Name} посещает лекцию.");
    }

    public void Contact()
    {
        Console.WriteLine($"{Name} можно связаться по Email: {Email} или Телефон: {PhoneNumber}");
    }
}

// Интерфейс для дополнительных возможностей
interface IResearch
{
    void ConductResearch();
    void PublishPaper();
}

// Класс BachelorStudent с простым наследованием
class BachelorStudent : Student
{
    public string Specialty { get; set; }
    public int Year { get; set; }
    public string InternshipCompany { get; set; }

    public BachelorStudent(string name, int age, string course, double gpa, string university, string specialty, int year, string internshipCompany, string email, string phoneNumber)
        : base(name, age, course, gpa, university, email, phoneNumber)
    {
        Specialty = specialty;
        Year = year;
        InternshipCompany = internshipCompany;
    }

    public override void Study()
    {
        Console.WriteLine($"{Name} учится на специальности {Specialty} в {Year} году обучения.");
    }

    public void ApplyForInternship()
    {
        Console.WriteLine($"{Name} подает заявку на стажировку в компанию {InternshipCompany}.");
    }
}

// Класс MasterStudent с простым наследованием и реализацией интерфейса IResearch
class MasterStudent : Student, IResearch
{
    public string ScientificAdvisor { get; set; }
    public string ResearchTopic { get; set; }
    public int Publications { get; set; }

    public MasterStudent(string name, int age, string course, double gpa, string university, string scientificAdvisor, string researchTopic, int publications, string email, string phoneNumber)
        : base(name, age, course, gpa, university, email, phoneNumber)
    {
        ScientificAdvisor = scientificAdvisor;
        ResearchTopic = researchTopic;
        Publications = publications;
    }

    public override void TakeExam()
    {
        Console.WriteLine($"{Name} сдает экзамен под руководством {ScientificAdvisor}.");
    }

    public void ConductResearch()
    {
        Console.WriteLine($"{Name} проводит исследование на тему '{ResearchTopic}'.");
    }

    public void PublishPaper()
    {
        Console.WriteLine($"{Name} публикует научную статью. Всего опубликованных статей: {Publications}.");
    }

    public void DefendThesis()
    {
        Console.WriteLine($"{Name} защищает магистерскую диссертацию.");
    }
}

// Класс PhDStudent с простым наследованием и реализацией интерфейса IResearch
class PhDStudent : Student, IResearch
{
    public string DissertationTopic { get; set; }
    public string Supervisor { get; set; }
    public int ResearchYears { get; set; }

    public PhDStudent(string name, int age, string course, double gpa, string university, string dissertationTopic, string supervisor, int researchYears, string email, string phoneNumber)
        : base(name, age, course, gpa, university, email, phoneNumber)
    {
        DissertationTopic = dissertationTopic;
        Supervisor = supervisor;
        ResearchYears = researchYears;
    }

    public override void Study()
    {
        Console.WriteLine($"{Name} работает над диссертацией на тему '{DissertationTopic}' под руководством {Supervisor}.");
    }

    public void ConductResearch()
    {
        Console.WriteLine($"{Name} проводит исследования уже {ResearchYears} лет.");
    }

    public void PublishPaper()
    {
        Console.WriteLine($"{Name} публикует статью по теме диссертации '{DissertationTopic}'.");
    }

    public void DefendThesis()
    {
        Console.WriteLine($"{Name} защищает свою диссертацию.");
    }
}

// Generic класс для работы со студентами
class StudentGroup<T> where T : Student
{
    private List<T> students = new List<T>();

    public void AddStudent(T student)
    {
        students.Add(student);
    }

    public void PrintAllStudents()
    {
        foreach (var student in students)
        {
            Console.WriteLine(student.GetInfo());
        }
    }

    public void ConductResearchForAll()
    {
        foreach (var student in students)
        {
            if (student is IResearch researcher)
            {
                researcher.ConductResearch();
            }
        }
    }
}
BachelorStudent bachelor = new BachelorStudent("Алиса", 20, "Компьютерные науки", 3.5, "Тюменский Государственный Университет", "Программная инженерия", 3, "Google", "alice@example.com", "123-456-7890");
MasterStudent master = new MasterStudent("Боб", 24, "Компьютерные науки", 3.8, "Тюменский Государственный Университет", "доктор Смит", "Машинное обучение", 5, "bob@example.com", "987-654-3210");
PhDStudent phd = new PhDStudent("Чарли", 28, "Компьютерные науки", 4.0, "Тюменский Государственный Университет", "Искусственный интеллект", "профессор Иванов", 4, "charlie@example.com", "555-666-7777");
// Работа с групповым generic классом StudentGroup
StudentGroup<Student> studentGroup = new StudentGroup<Student>();
studentGroup.AddStudent(bachelor);
studentGroup.AddStudent(master);
studentGroup.AddStudent(phd);
Console.WriteLine("Информация о всех студентах:");
studentGroup.PrintAllStudents();
Console.WriteLine("\nИсследования всех студентов:");
studentGroup.ConductResearchForAll();
```
